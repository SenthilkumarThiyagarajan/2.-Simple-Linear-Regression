```python
import pandas as pd # Here we are importing pands library. Then Instead of using pandas in application we can use pd
```

```python
dataset=pd.read_csv("Salary_Data.csv") #Here we are reading Salary_data.csv and assigning to dataset variable
```

```python
dataset # we are printing dataset, Upto this Data Collection Completed
```

```python
#Data preprocessing started
independent = dataset[["YearsExperience"]]# independent is input variable

independent #Here we are printing input variable
```

```python
dependent = dataset[["Salary"]] # Dependent is output variable
dependent # We are printing 

#Input/Output split done
```

```python
#Here we are spliting  Train, Testset
from sklearn.model_selection import train_test_split # All machingelanguage libraries are stored in sklearn file.Here we import train test split method
X_train,X_test,Y_train,Y_test = train_test_split(independent, dependent, test_size=0.30, random_state=0) #Passing independent(input), dependent(output),test_size,randomsate

#Train independent,train dependent values are stored in X_train, Y_train
#Test independent,train dependent values are stored in X_test, Y_test

#Train and Test data split done
```

```python
#Model Creation using train data

from sklearn.linear_model import LinearRegression 
regressor = LinearRegression() #Assigning all function of LinearRegression toregressor 
regressor.fit(X_train,Y_train) #We are passing X_train, Y_train as parameter to fit method
```

```python
weight = regressor.coef_#This is used to get slope(m) or weight(w)
```

```python
weight #printing weight 
```

```python
bias = regressor.intercept_#This isused to get bias c or b
bias#printing bias
```

```python
y_pred = regressor.predict(X_test)#Here we are passing our test set to model to pedict y
```

```python
y_pred
```

```python
# Validating Parameters - Evalution metrics
from sklearn.metrics import r2_score
r_score = r2_score(Y_test,y_pred) #We are passing our test set to r2_score
r_score#printing ,If value is near to 1,then that is good model, if it is near to zero,then it is bad model

# Evalution metrics done
```

```python
#Save the model
import pickle #pickle is used to save the model
filename="finalized_model_linear.sav" # Here we are giving file name
```

```python
pickle.dump(regressor,open(filename,'wb')) # Here we are writing our model in this file, wb is write binary
```

```python
#Deployment
loaded_model = pickle.load(open("finalized_model_linear.sav",'rb')) # Here we are opening our model and assign to loaded_model variable
result=loaded_model.predict([[15]])#Here we are passing 15 as year parameter to predict method
result# Printing our values
```
